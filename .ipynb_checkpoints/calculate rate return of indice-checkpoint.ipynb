```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickers = ['^JKLQ45', '^JKSE', '^JKII']
my_data = pd.DataFrame()
for t in tickers:
    my_data[t] = wb.DataReader(t, "yahoo", start='2013-3-05')['Adj Close']
    
my_data.head()
```

##### Normalization to 100

```python
my_data.iloc[0]
```

```python
(my_data / my_data.iloc[0] * 100).plot(figsize=(15,8))
```

```python
my_data.plot(figsize=(16, 10))
```

### Calculating the return of indice

```python
returns = (my_data / my_data.shift(1)) -1 
returns.tail()
```

```python
annual_returns = returns.mean() * 250 
annual_returns
```
